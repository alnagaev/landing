```python
import requests
import urllib
user_agent = {'User-agent': 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:71.0) Gecko/20100101 Firefox/71.0'}
success_ip = []
price_list = {}
count = 0
success = False

for link in links[0:3]:
    
        
    try:
        r = requests.get(link,
                        headers = user_agent, proxies={
                      "http"  : ips_list[count],
                      "https" : ips_list[count],
                      "ftp"   : ips_list[count]
                    },  timeout=(6, 60) )

        if r.status_code != 200:
            success = False
            count +=1
            pass

        else:
            soup = BeautifulSoup(r.text)
            price = soup.select('.js-item-price')[0]['content']
            price_list[] = price
            success_ip.append(ips_list[count])
            
    except:
        count +=1
        pass

        
print(price_list)
```
